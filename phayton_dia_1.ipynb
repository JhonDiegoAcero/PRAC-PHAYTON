INTRODUCCION A NUMPY: ES UNA BIBLIOTECA FUNDAMENTAL PARA MANIPULAR DE ARRAYS Y OPERACIONES MATEMATICAS EN PHYTON

```python
import numpy as pn
#crear un array numpy
array = pn.array([1,2,3,4,5])
#realizar operaciones basicas
suma=pn. sum(array)
media=pn.mean(array)
print("La suma es",suma)
print("La media es",media)
```

EJERCICIO 2:
CARGAR Y VISUALIZAR UN CONJUNTO DE DATOS CON PANDAS Y MATPLOTLIB

```python
import pandas as pd
import matplotlib.pyplot as plt

#cargar un conjunos de datos 'iris'
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',header=None,names=['sepal_length','sepal_width','petal_length','petal_width','class'])
#mostrar las primeras filas del dataset
print(iris.head())
#visualizar la distribucion de la dataset
plt.hist(iris['petal_length'],bins=20, color='blue',edgecolor='red')
plt.title('Distribucion de la longuitud de los Petalos')
plt.xlabel('Longuitud de los Petalos')
plt.ylabel('Frecuencia')
plt.show()
```
